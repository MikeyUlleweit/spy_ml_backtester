```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("SPY_Features.csv")
df["Date"] = pd.to_datetime(df["Date"])

plt.figure(figsize=(14, 6))

# Plot MACD and RSI 
plt.plot(df["Date"], df["macd_scaled"], label="MACD (Scaled)", color="blue", alpha=0.7)
plt.plot(df["Date"], df["rsi_scaled"], label="RSI (Scaled)", color="purple", alpha=0.7)

# Add targes green for target=1/up, red for target=0/down
plt.scatter(df[df["target"] == 1]["Date"], df[df["target"] == 1]["macd_scaled"],
            color="green", label="Up (Target = 1)", s=10)
plt.scatter(df[df["target"] == 0]["Date"], df[df["target"] == 0]["macd_scaled"],
            color="red", label="Down (Target = 0)", s=10)

# Plot formatting
plt.title("MACD and RSI (Scaled) with Market Direction Target")
plt.xlabel("Date")
plt.ylabel("Scaled Indicator Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
