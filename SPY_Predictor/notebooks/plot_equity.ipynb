```python
import pandas as pd
import matplotlib.pyplot as plt

# Load equity curve
df = pd.read_csv("equity_curve.csv")

# Hardcoded values from C++ output
initial_equity = 10000.0
final_equity = 14055.9
total_return_pct = 40.5593
total_trades = 2639            # from C++
win_rate = 54.1493             # from C++
total_days = len(df)           # actual number of equity entries

# Plot
plt.figure(figsize=(10, 5))
plt.plot(df["day"], df["equity"], label="Equity Curve", linewidth=2)
plt.title("SPY Strategy Backtest: Equity Curve", fontsize=14)
plt.xlabel("Day", fontsize=12)
plt.ylabel("Account Value ($)", fontsize=12)
plt.grid(True)
plt.legend()

# Annotate key metrics
plt.figtext(0.15, 0.80, f"Initial Equity: ${initial_equity:,.2f}")
plt.figtext(0.15, 0.75, f"Final Equity: ${final_equity:,.2f}")
plt.figtext(0.15, 0.70, f"Total Return: {total_return_pct:.2f}%")
plt.figtext(0.15, 0.65, f"Total Trades: {total_trades}")    # FIXED
plt.figtext(0.15, 0.60, f"Win Rate: {win_rate:.2f}%")
plt.figtext(0.15, 0.55, f"Total Days: {total_days}")

plt.tight_layout()
plt.savefig("equity_curve_plot.png")
plt.show()
```
